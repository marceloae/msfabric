### Execução Paralela com DAG

```python
DAG = {
    "activities": [
        {
            "name": "Notebook1",
            "path": "Notebook1",
            "timeoutPerCellInSeconds": 90
        },
        {
            "name": "Notebook2",
            "path": "Notebook2",
            "timeoutPerCellInSeconds": 90
        },
        {
            "name": "Notebook3",
            "path": "Notebook3",
            "timeoutPerCellInSeconds": 90
        }
    ]
}
mssparkutils.notebook.runMultiple(DAG, {"displayDAGViaGraphviz": True})
```
